```python
import numpy as np
import pandas as pd
import re
import emoji
```

**CSV File Upload**

```python
from google.colab import files
data_frame = files.upload()
```

```python
df = pd.read_csv('Dataset for Annotation.csv', encoding='utf-8-sig')
print(df.columns)
print(df.shape)
print("Null Values Check:")
print(df.isnull().sum())
```

```python
print(df.head(25))
print(df.tail(25))
```

**Dataset Cleaning Function**

```python
TEXT_COLUMN = 'Comment'

def clean_bangla_text(Comment):

  Comment = str(Comment)

  #Removing URLs
  Comment = re.sub(r'http\S+|www\S+', '', Comment)

  #Removu=ing English words
  Comment = re.sub(r'[A-Za-z]+', ' ', Comment)

  #Removing English numbers
  Comment = re.sub(r'[0-9]+', ' ', Comment)

  #Eliminating Emojis
  Comment = emoji.replace_emoji(Comment, replace='')

  #Keeping only the Bangla characters
  Comment = re.sub(r'[^ঀ-৿\s।?!]', ' ', Comment)

  #Eliminating extra spaces
  Comment = re.sub(r'\s+', ' ', Comment).strip()

  return Comment


df['cleaned_text'] = df[TEXT_COLUMN].apply(clean_bangla_text)

#Eliminating empty rows
df = df[df['cleaned_text'].str.strip() != '']

#Removing Duplicates
df.drop_duplicates(subset=['cleaned_text'], inplace=True)

#Index reset
df.reset_index(drop=True, inplace=True)

print(df[['cleaned_text']].head(20))
```

```python
print(df[['cleaned_text']].tail(50))
```

```python
output_file = '/content/cleaned_dataset.csv'

df.to_csv(output_file, index=False, encoding='utf-8-sig')

print(f"\nCleaned dataset saved to: {output_file}")
```

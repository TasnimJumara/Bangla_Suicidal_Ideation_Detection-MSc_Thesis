**Applying both Cohen's Kappa and Weighted Cohen's Kappa**

```python
from sklearn.metrics import cohen_kappa_score

df = pd.read_csv("Output File_Annotation Checking of Tasnim_Joy.csv")

#Calculation of Cohen's Kappa
kappa = cohen_kappa_score(
    df['Tasnim_Annot'],
    df['Joy_Annot']
)

print("The Cohen's Kappa score is:", round(kappa, 4))

#Calculation of Weighted Cohen's Kappa
kappa = cohen_kappa_score(
    df['Tasnim_Annot'],
    df['Joy_Annot'],
    weights='quadratic'
)

print("The weighted Cohen's Kappa is:", round(kappa, 4))
```

**Detecting Mismatched Data**

```python
mismatch_df = df[
    df['Tasnim_Annot'] != df['Joy_Annot']
]

print("Total mismatched sample is:", len(mismatch_df))

print("\nThe mismatched samples are:\n")
print(mismatch_df)
```

**Saving Mismatched Data**

```python
mismatch_df.to_csv(
    "Mismatched_Samples of Tasnim_Joy.csv",
    index=False,
    encoding='utf-8-sig'
)

print("Mismatch file has been saved successfully.")
```
